```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import requests
import time
import urllib
```

```python
# file to load 
file_to_load= 'Resources/cities.csv'
```

```python
# read in data frame 
file_df= pd.read_csv(file_to_load)
file_df.head()
```

```python
html_table = file_df.to_html()
html_table
```

```python
file_df.to_html('table.html')
```

```python

import matplotlib.pyplot as plt
# Extract relevant fields from the data frame
lats = file_df["Lat"]
max_temps = file_df["Max Temp"]
humidity = file_df["Humidity"]
cloudiness = file_df["Cloudiness"]
wind_speed = file_df["Wind Speed"]

plt.scatter(lats, 
            max_temps,
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Cities")

# Incorporate the other graph properties
plt.title("City Latitude vs. Max Temperature (%s)" % time.strftime("%x"))
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

# save picture
plt.savefig('fig1.png')


# Show plot
plt.show()

```
